```python
import tkinter as tk
from tkinter import messagebox
from tkinter import *

import os
import shutil

window=tk.Tk()
window.title("Data Syncer")
window.geometry("400x400")


#To enter source path
sp= tk.Label(text="Enter source path: ")
sp.grid(column='1',row='1')

esp= tk.Entry()
esp.grid(column='2',row='1')

#To enter destination path
dp= tk.Label(text="Enter destination path: ")
dp.grid(column='1',row='2')

edp= tk.Entry()
edp.grid(column='2',row='2')

#To enter the file category(type)
fc= tk.Label(text="Enter file category: ")
fc.grid(column='1',row='3')

efc= tk.Entry()
efc.grid(column='2',row='3')

def Manual():
    messagebox.showinfo( "File Category Manual!", "1- Photos, 2- Text, 3- Songs")

B = tk.Button(text ="File Category Manual!", command = Manual)

B.grid(column="3",row="3")



def Input():
    op= Entry.get(esp)
    dp= Entry.get(edp)
    choice=int(Entry.get(efc))
    return op,dp,choice
def Backup():
    i,o,ch=Input()
    if(ch==1):
        l=o+"\\Photos"
        files_source=os.listdir(i)
        if(os.path.isdir(l)==False):
            files_destin=os.mkdir(l)
        files_destin=os.listdir(l)
    elif(ch==2):
        l=o+"\\Texts"
        files_source=os.listdir(i)
        if(os.path.isdir(l)==False):
            files_destin=os.mkdir(l)
        files_destin=os.listdir(l)
    elif(ch==3):
        l=o+"\\Songs"
        files_source=os.listdir(i)
        if(os.path.isdir(l)==False):
            files_destin=os.mkdir(l)
        files_destin=os.listdir(l)
    else:
        print("Wrong choice")
    if(ch==1 or ch==2 or ch==3):
        imt=[]
        if(files_destin==[]):
            for f in files_source:
                ff=os.path.join(i,f)
                imt.append(os.path.getmtime(ff))
                if(os.path.isfile(ff)):
                    shutil.copy(ff,l)
            print("Files backed up successfully")
        else:
            for f in files_destin:
                ff=os.path.join(l,f)
                imt.append(os.path.getmtime(ff))
            k=0
            for f in files_source:
                if(f in files_destin):
                    ff=os.path.join(l,f)
                    mt=os.path.getmtime(ff)
                    if(mt!=imt[k]):
                        if(os.path.isfile(ff)):
                            shutil.copy(ff,l)
                    k=k+1
                else:
                    ff=os.path.join(i,f)
                    if(os.path.isfile(ff)):
                        shutil.copy(ff,l)
                    k=k+1
            print("Files backed up successfully")
    else:
        print("Backup failed")
        
        
def message():
    messagebox.showinfo( "Congratulations!", "Your files have been backed up!")
#For Run! button
run= tk.Button(text="Run!", command=Backup)
run.grid(column='2',row='4')

window.mainloop()
```
